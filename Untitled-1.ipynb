```python
# FUNÇÃO PARA DISTRIBUIR TRIMESTRAL PARA MENSAL
import pandas as pd
import os
def distribute_trimestral_to_monthly(df):
    result = []
    #df= pd.read_excel(os.getcwd() + "\\bacen_consorcio_UF.xlsx")
    for _, row in df.iterrows():
        year_month = row['#Data_base']
        year = int(str(year_month)[:4])
        month = int(str(year_month)[4:])
          
        # MAPEAMENTO DE TRIMESTRE PARA MESES CORRESPONDENTES
        if month == 3:
            months = [(year, 1), (year, 2), (year, 3)]
        elif month == 6:
            months = [(year, 4), (year, 5), (year, 6)]
        elif month == 9:
            months = [(year, 7), (year, 8), (year, 9)]
        elif month == 12:
            months = [(year, 10), (year, 11), (year, 12)]
        
        value = row['Quantidade_de_adesões_no_trimestre']
        base_value = value // 3
        remainder = value % 3
        
        # DISTRIBUI OS VALORES IGUALMENTE E AJUSTA O RESTO
        for i in range(3):
            month_year, month = months[i]
            # ADICIONA 1 AOS PRIMEIRO 'REMAINDER' MES PARA DISTRIBUIR O RESTO
            qtd_month = base_value + (1 if i < remainder else 0)
            
            result.append({
                'Data': str(month_year) + "/" + str(month) + '/01',
                'Quantidade_de_adesões_no_mês': qtd_month
            })
    df = pd.DataFrame(result)
    df['Data'] = pd.to_datetime(df['Data'], format='%Y/%m/%d')
    df_grouped = df.sort_values(by='Data').reset_index(drop=True)
    df_grouped = df.groupby(df['Data']) \
                    .agg({'Quantidade_de_adesões_no_mês': 'sum'}) \
                    .reset_index()
    df_grouped['Quantidade_de_adesões_no_mês'] = df_grouped['Quantidade_de_adesões_no_mês'].astype(int)
    return df_grouped
# Aplicando a função ao DataFrame
```

```python
import pandas as pd
import numpy as np
import os
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df_financiamento = pd.read_excel(os.getcwd() + "\\sheets\\bacen_financiamento.xlsx")
df_consorcio = pd.read_excel(os.getcwd() + "\\sheets\\bacen_consorcios_acumulado.xlsx")
df_consorcio = df_consorcio.rename(columns={'Data_base': 'Data'})

df_financiamento['Data'] = pd.to_datetime(df_financiamento['Data'], format='%d/%m/%Y')
df_consorcio['Data'] = pd.to_datetime(df_consorcio['Data'], format='%d/%m/%Y')
df_financiamento['Qtd_aquisições'] = df_financiamento['Qtd_aquisições'].replace(',', '').astype(float)


df_financiamento['Data_num'] = df_financiamento['Data'].map(pd.Timestamp.toordinal)
df_consorcio['Data_num'] = df_consorcio['Data'].map(pd.Timestamp.toordinal)
data_financiamento = df_financiamento[['Data_num', 'Qtd_aquisições']]
data_consorcio = df_consorcio[['Data_num', 'Quantidade_de_adesões_no_mês']]
# NORMALIZAR OS DADOS
scaler_fin = StandardScaler()
data_financiamento_normalized = scaler_fin.fit_transform(data_financiamento)

scaler_cons = StandardScaler()
data_consorcio_normalized = scaler_cons.fit_transform(data_consorcio)

# EXECUTAR O K-MEANS PARA DETERMINAR O MELHOR NÚMERO DE CLUSTERS

def optimal_clusters(data):
    sse = []
    k_range = range(1, 11)
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=0)
        kmeans.fit(data)
        sse.append(kmeans.inertia_)
    return sse

sse_fin = optimal_clusters(data_financiamento_normalized)
sse_cons = optimal_clusters(data_consorcio_normalized)

plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plt.plot(range(1, 11), sse_fin, marker='o', label='Financiamento')  
plt.xlabel('Número de Clusters', fontsize=12)
plt.ylabel('Soma dos Quadrados das Distâncias (SSE)', fontsize=12)
plt.title('Método do Cotovelo - Financiamento', fontsize=14)
plt.legend(loc='best', fontsize=12) 

plt.subplot(1, 2, 2)
plt.plot(range(1, 11), sse_cons, marker='o', label='Consórcio')  
plt.xlabel('Número de Clusters', fontsize=12)
plt.ylabel('Soma dos Quadrados das Distâncias (SSE)', fontsize=12)
plt.title('Método do Cotovelo - Consórcio', fontsize=14)
plt.legend(loc='best', fontsize=12)  
plt.tight_layout()
plt.show()

# ESCOLHER O NÚMERO DE CLUSTERS (POR EXEMPLO, 3 PARA AMBOS OS CASOS) E AJUSTAR O MODELO K-MEANS
kmeans_fin = KMeans(n_clusters=3, random_state=0)
df_financiamento['Cluster'] = kmeans_fin.fit_predict(data_financiamento_normalized)
kmeans_cons = KMeans(n_clusters=3, random_state=0)
df_consorcio['Cluster'] = kmeans_cons.fit_predict(data_consorcio_normalized)
```

```python
import pandas as pd
import numpy as np


credito = 242077.21  
prazo = 200  
taxa_admin = 19.43 / 100  
fundo_reserva = 3 / 100 
incc_ano = 3.32 / 100  

custo_inicial_consorcio = credito * (1 + taxa_admin + fundo_reserva)
meses = np.arange(1, prazo + 1)

parcelas_consorcio = []
saldo_devedor_consorcio = []
taxa_admin_list = []
fundo_reserva_list = []

parcela_inicial_consorcio = custo_inicial_consorcio / prazo  
parcela_consorcio_anual = parcela_inicial_consorcio  

saldo_consorcio = custo_inicial_consorcio

for mes in meses:
    saldo_consorcio -= parcela_consorcio_anual
    if mes % 12 == 0: 
        saldo_consorcio *= (1 + incc_ano)  
        parcela_consorcio_anual = float(saldo_consorcio / (prazo - mes))  
    parcelas_consorcio.append(round(parcela_consorcio_anual, 2))
    saldo_devedor_consorcio.append(round(saldo_consorcio, 2))
    taxa_admin_list.append(round(parcela_consorcio_anual * taxa_admin, 2))
    fundo_reserva_list.append(round(parcela_consorcio_anual * fundo_reserva, 2))

total_taxa_admin = sum(taxa_admin_list)
total_fundo_reserva = sum(fundo_reserva_list)

print(f"Total Taxa de Administração: {total_taxa_admin}")
print(f"Total Fundo de Reserva: {total_fundo_reserva}")

df_simulacao_consorcio = pd.DataFrame({
    'Mês': meses,
    'Parcela Consórcio': parcelas_consorcio,
    'Saldo Devedor Consórcio': saldo_devedor_consorcio,
    'Taxa de Administração': taxa_admin_list,
    'Fundo de Reserva': fundo_reserva_list
})

resumo_consorcio = {
    'Consórcio': {
        'Custo Inicial': round(custo_inicial_consorcio, 2),
        'Parcela Inicial': round(parcela_inicial_consorcio, 2),
        'Custo Final (com INCC)': round(sum(parcelas_consorcio) + total_taxa_admin + total_fundo_reserva, 2),
        'Total Taxa de Administração': round(total_taxa_admin, 2),
        'Total Fundo de Reserva': round(total_fundo_reserva, 2)
    }
}

print("\nResumo Final do Consórcio:")
print(resumo_consorcio)
df__ano = df_simulacao_consorcio[df_simulacao_consorcio['Mês'] % 12 == 0]
df__ano.to_excel(os.getcwd() + "\\consorcio_detalhado.xlsx", index=False)
```

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
from matplotlib.cm import viridis

df_financiamento = pd.read_excel(os.getcwd() + "\\financiamento_detalhado.xlsx")
df_consorcio = pd.read_excel(os.getcwd() + "\\consorcio_detalhado.xlsx")

# Ajustar os dados
df_consorcio['Fundo de Reserva'] = df_consorcio['Fundo de Reserva'].replace(".", "").replace(',', '.').astype(float)
df_financiamento['Juros'] = df_financiamento['Juros'].replace(".", "").replace(',', '.').astype(float)
df_financiamento['Parcela Financiamento (PRICE)'] = df_financiamento['Parcela Financiamento (PRICE)'].replace(".", "").replace(',', '.').astype(float)

# Dados ajustados (com soma da Taxa de Administração + Fundo de Reserva para Consórcio)
df_consorcio['Taxa + Fundo'] = df_consorcio['Taxa de Administração'] + df_consorcio['Fundo de Reserva']

# Médias para linhas horizontais
media_custo_financiamento = df_financiamento['Parcela Financiamento (PRICE)'].mean()
media_custo_consorcio = df_consorcio['Parcela Consórcio'].mean()
print(media_custo_consorcio)

# Gráfico de barras
x = df_financiamento['Número Parcela']
width = 3  # Largura das barras

fig, ax = plt.subplots(figsize=(14, 7))
bars1 = ax.bar(x - width/2, df_financiamento['Parcela Financiamento (PRICE)'], width, label='Custo Parcela (Financiamento)', color=viridis(0.6), alpha=0.8)
bars2 = ax.bar(x + width/2, df_consorcio['Parcela Consórcio'], width, label='Custo Parcela (Consórcio)', color=viridis(0.1), alpha=0.8)

# Linhas médias
ax.axhline(media_custo_financiamento, color='green', linestyle='--', linewidth=1.5, label=f'Média Custo Parcela Financiamento: {media_custo_financiamento:.2f}')
ax.axhline(media_custo_consorcio, color='darkviolet', linestyle='--', linewidth=1.5, label=f'Média Custo Parcela Consórcio: {media_custo_consorcio:.2f}')

# Valores no topo das barras
for bar in bars1:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height + 0.01, f'{height:.2f}', ha='center', va='bottom', fontsize=10)
for bar in bars2:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height + 0.01, f'{height:.2f}', ha='center', va='bottom', fontsize=10)

# Customização do gráfico
ax.set_title('Custo Total: Financiamento vs Consórcio', fontsize=16, pad=20)
ax.set_xlabel('Número de Parcelas', fontsize=14)
ax.set_ylabel('Valor (R$)', fontsize=14)
ax.legend(fontsize=12)
ax.grid(True, linestyle='--', alpha=0.7)
plt.xticks(x, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
